```python
from PyDictionary import PyDictionary
dictionary=PyDictionary()
print (dictionary.meaning("researcher's"))
def get_text(position, occupation):
    lst = position.split() + occupation.replace('& ','').split()
    ans = []
    for l in lst:
        d = dictionary.meaning(l)
        try:
            for el in d:
                ans.append(d[el][0])
        except:
            ans.append(l)
    return ' '.join(ans)
```
